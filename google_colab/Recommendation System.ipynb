### Split Train Test Set

```python
# split train and test set by length


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_sequence_song, all_sequence_song, test_size=0.2, random_state=42, stratify=[len(i) for i in all_sequence_song ]); X_train= sample(all_sequence_song,int(len(all_sequence_song)*0.95))
```

```python
def create_overlap_dataset(all_sequences, train= True):
  if train:
    X, y = [], []
    for seq in all_sequences:
        if len(seq) > 3:
            for i in range(3, len(seq)):
                X.append(seq[:i])
                y.append(seq[i])
        else:
            # If sequence length is less than or equal to 3, this else block may not be correctly handling the sequences.
            # Typically, you would want to handle this case differently or exclude such short sequences.
            X.append(seq[:3])
            y.append(seq[3])
  else:
    X, y = [], []
    for seq in all_sequences:
        X.append(seq[:len(seq)-1])
        y.append(seq[len(seq)-1])

  return X, y
```

### Define Dataloader and Model

```python
# define dataset & dataloader

def pad_seq(arr, max_len=19):
  new_arr= np.array(list(arr) + [0]*(max_len-len(arr)))
  return new_arr

class MusicDataset(Dataset):
    def __init__(self, user_sequences, next_sequences):
        self.user_sequences = user_sequences
        self.next_sequences = next_sequences
        assert len(user_sequences) == len(next_sequences)

    def __len__(self):
        return len(self.user_sequences)

    def __getitem__(self, idx):
        # Assuming the sequence is already encoded as integers
        sequence = pad_seq(self.user_sequences[idx])
        next_ = self.next_sequences[idx]

        # Input is all tracks except the last, target is the last track
        return torch.tensor(sequence), torch.tensor(next_)



# Assume `user_sequences` is a collection of sequences, one per user
train_dataset = MusicDataset(create_overlap_dataset(X_train, True)[0], create_overlap_dataset(X_train, True)[1])
test_dataset = MusicDataset(create_overlap_dataset(X_test, True)[0], create_overlap_dataset(X_test, True)[1])
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=True)
```

```python
print('num train: ', len(train_loader.dataset))
print('num test: ',len(test_loader.dataset))
```

```python
# define model
device = "cuda:0"

class MusicRecommender(nn.Module):
    def __init__(self, num_tracks, embedding_dim, hidden_dim):
        super(MusicRecommender, self).__init__()
        self.embedding = nn.Embedding(num_tracks, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_tracks)

    def forward(self, sequences):
        embeddings = self.embedding(sequences)
        lstm_out, _ = self.lstm(embeddings)
        logits = self.fc(lstm_out[:, -1, :])
        return logits

num_tracks = len(label_encoder.classes_)  # Number of unique tracks in dataset
embedding_dim = 512  # Size of the embedding layer
hidden_dim = 1024  # Size of the LSTM's hidden layer

model = MusicRecommender(num_tracks, embedding_dim, hidden_dim).to(device)
```

### Training and Testing

```python
from sklearn.metrics import accuracy_score

def validate_model(model, val_loader):
    model.eval()
    all_predictions = []
    all_targets = []
    with torch.no_grad():
        for sequences, targets in val_loader:
            predictions = model(sequences.to(device))
            _, predicted_indexes = torch.max(predictions, 1)
            all_predictions.extend(predicted_indexes.cpu().numpy())
            all_targets.extend(targets.numpy())
    val_accuracy = accuracy_score(all_targets, all_predictions)
    return val_accuracy

criterion = nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

num_epochs = 15  # Number of epochs to train for

best_val_accuracy = 0
best_model_state = None

for epoch in range(num_epochs):
    model.train()
    all_train_predictions = []
    all_train_targets = []
    for sequences, targets in train_loader:
        optimizer.zero_grad()
        predictions = model(sequences.to(device))
        loss = criterion(predictions, targets.to(device))
        loss.backward()
        optimizer.step()

        _, predicted_indexes = torch.max(predictions, 1)
        all_train_predictions.extend(predicted_indexes.cpu().numpy())
        all_train_targets.extend(targets.numpy())

    train_accuracy = accuracy_score(all_train_targets, all_train_predictions)
    val_accuracy = validate_model(model, test_loader)

    print(f'Epoch {epoch+1}/{num_epochs}, Train Accuracy: {train_accuracy}, Validation Accuracy: {val_accuracy}')

    # Save model if validation accuracy has increased
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        best_model_state = model.state_dict().copy() # copy state dict, not a reference
        torch.save(model.state_dict(), '/content/drive/MyDrive/Colab Notebooks/BigData/best_model_acc.pth') # Save the best model's state dict
        print('Save model !')

# After training, load the best model state
model.load_state_dict(best_model_state)
```

```python
# define model
import torch.nn as nn
import torch
from random import sample
import numpy as np
class MusicRecommender(nn.Module):
    def __init__(self, num_tracks, embedding_dim, hidden_dim):
        super(MusicRecommender, self).__init__()
        self.embedding = nn.Embedding(num_tracks, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, num_tracks)

    def forward(self, sequences):
        embeddings = self.embedding(sequences)
        lstm_out, _ = self.lstm(embeddings)
        logits = self.fc(lstm_out[:, -1, :])
        return logits

num_tracks = len(label_encoder.classes_)  # Number of unique tracks in dataset
embedding_dim = 512  # Size of the embedding layer
hidden_dim = 1024  # Size of the LSTM's hidden layer

model = MusicRecommender(num_tracks, embedding_dim, hidden_dim)
model.load_state_dict(torch.load('/content/drive/MyDrive/Colab Notebooks/BigData/best_model_acc.pth'))
```

```python
def pad_seq(arr, max_len=19):
  new_arr= np.array(list(arr) + [0]*(max_len-len(arr)))
  return new_arr
def recommend_next_track(model, top_song, user_sequence, track_id_encoder=label_encoder):
    #
    if len(user_sequence)<3:
      return sample(top_song, 1)[0]
    elif len(user_sequence)>19:
      user_sequence= user_sequence[-20:]
    model.eval()

    with torch.no_grad():
        sequence_tensor = torch.tensor(pad_seq(track_id_encoder.transform(user_sequence))).unsqueeze(0)  # Add batch dimension
        logits = model(sequence_tensor)
        recommended_track_id = logits.argmax(dim=1).item()  # Get the track with the highest score
    return track_id_encoder.inverse_transform([recommended_track_id])[0]

```

```python
import gradio as gr

top_song= list(all_song.copy())
user_sequence= []
def recommend_and_track(song):
    # Access the user_sequence list defined outside of the function
    # global user_sequence

    # Append the selected song to the user_sequence list
    user_sequence.append(song)

    # Get recommendations based on the selected song
    recommendations = recommend_next_track(model, top_song, user_sequence, label_encoder)

    # Return both the recommendations and the updated user_sequence list
    return recommendations, '\n'.join(user_sequence)
```

```python
# Define the Gradio interface
iface = gr.Interface(
    fn=recommend_and_track,
    inputs=gr.Dropdown(choices=all_song, label="Select a Song"),
    outputs=[
        gr.Textbox(label="Recommended Songs"),
        gr.Textbox(label="Previously Selected Songs"),
    ]
)

# Launch the interface
iface.launch()
```
